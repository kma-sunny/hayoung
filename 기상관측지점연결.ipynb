### 유클리디안 거리 

### dst=distance.euclidean(a,b)

```python
for i in range(md.shape[0]):
    #print(md.iloc[i][['lat','long']])
    wantedPoint=md.iloc[i][['lat','long']]
    minDist=100000
    for j in range(spot_info.shape[0]):
        #print(spot_info.iloc[j])
        point=spot_info.iloc[j][['Latitude','Longitude']]
        minDist=min(minDist, distance.euclidean(point,wantedPoint))
    print(minDist)
```

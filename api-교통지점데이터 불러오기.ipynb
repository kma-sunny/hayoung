```python
# Extract and output tags of interest
```

```python
spot_num=[]
spot_nm=[]
grs80tm_x=[]
grs80tm_y=[]
```

```python
#while endnumber<=1000:
url=url.format(key=serviceKey, startIndex=1, endIndex=1000)
req=requests.get(url)
html=req.text
soup=BeautifulSoup(html,'html.parser')

#print(soup)
spot_num=pd.Series(soup.find_all('spot_num'))
spot_nm=pd.Series(soup.find_all('spot_nm'))
grs80tm_x=pd.Series(soup.find_all('grs80tm_x'))
grs80tm_y=pd.Series(soup.find_all('grs80tm_y'))

idx=pd.Series(list(range(len(spot_num))))
```

```python
nm=list(soup.find_all('spot_nm'))
temp=[]
for da in nm:
    temp.append((str(da).split("<spot_nm>"))[1].split("</spot_nm>")[0])
```

```python
spot_nm=pd.Series(temp)
spot_nm
```

```python
nm=list(soup.find_all('spot_num'))
temp=[]
for da in nm:
    temp.append((str(da).split("<spot_num>"))[1].split("</spot_num>")[0])
    
spot_num=pd.Series(temp)
spot_num    
```

```python
nm=list(soup.find_all('grs80tm_x'))
temp=[]
for da in nm:
    temp.append((str(da).split("<grs80tm_x>"))[1].split("</grs80tm_x>")[0])
    
grs80tm_x=pd.Series(temp)
grs80tm_x 
```

```python
nm=list(soup.find_all('grs80tm_y'))
temp=[]
for da in nm:
    temp.append((str(da).split("<grs80tm_y>"))[1].split("</grs80tm_y>")[0])
    
grs80tm_y=pd.Series(temp)
grs80tm_y 
```

```python
spot_num=pd.DataFrame(data={'spot_num':spot_num, 'index': idx},index=range(len(spot_num)))
spot_nm=pd.DataFrame(data={'spot_nm':spot_nm, 'index':idx}, index=range(len(spot_nm)))
grs80tm_x=pd.DataFrame(data={'grs80tm_x':grs80tm_x,'index':idx}, index=range(len(grs80tm_x)))
grs80tm_y=pd.DataFrame(data={'grs80tm_y':grs80tm_y,'index':idx}, index=range(len(grs80tm_y)))
```

```python
print(spot_num.shape)
print(spot_nm.shape)
print(grs80tm_x.shape)
print(grs80tm_y.shape)
```

## Merging data

```python
df=pd.merge(spot_num, spot_nm, how='left')
df=pd.merge(df,grs80tm_x, how='left')
df=pd.merge(df,grs80tm_y,how='left')
df
```

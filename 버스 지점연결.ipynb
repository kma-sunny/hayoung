# 정류소 X, Y좌표 정보 저장

```python
stopName_XY={}
for i in range(busStop.shape[0]):
    li=[]
    stopName_XY[busStop['정류소명'].iloc[i]]=busStop[['Y좌표','X좌표']].iloc[i]
```

```python
len(stopName_XY)
```

```python
stopName_XY.values()
```

```python
for i in range(busUser_info.shape[0]):

    if busUser_info['bus_sta_nm'].iloc[i] in list(stopName_XY.keys()):
        
        busUser_info['lati'].iloc[i]=stopName_XY[busUser_info['bus_sta_nm'].iloc[i]]['Y좌표']
        busUser_info['long'].iloc[i]=stopName_XY[busUser_info['bus_sta_nm'].iloc[i]]['X좌표']
    
        
#         print(stopName_XY[busUser_info['bus_sta_nm'].iloc[i]]['Y좌표'])
#         print(stopName_XY[busUser_info['bus_sta_nm'].iloc[i]]['X좌표'])
    
        
    
    
```

```python
for i in range(busUser_info.shape[0]):
    if len(busStop[busStop['정류소명']==busUser_info.iloc[i]['bus_sta_nm']]['Y좌표']) > 0 :  
        busUser_info['lati'].iloc[i]=busStop[busStop['정류소명']== busUser_info.iloc[i]['bus_sta_nm']]['Y좌표'].iloc[0]
        #print(busStop[busStop['정류소명']== busUser_info.iloc[i]['bus_sta_nm']]['Y좌표'].iloc[0])
    else: # 위치 정보 없을때
        busUser_info['lati'].iloc[i]= 0
        
    if len(busStop[busStop['정류소명']==busUser_info.iloc[i]['bus_sta_nm']]['X좌표']) > 0:
        busUser_info['long'].iloc[i]=busStop[busStop['정류소명']== busUser_info.iloc[i]['bus_sta_nm']]['X좌표'].iloc[0]
        #print(busStop[busStop['정류소명']== busUser_info.iloc[i]['bus_sta_nm']]['X좌표'].iloc[0])
    else: # 위치 정보 없을때
        busUser_info['long'].iloc[i] = 0
```
